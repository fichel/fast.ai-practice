```python
import json
import random

import fastai
import torch
from fastai.text.all import *
from fastcore.all import *
from transformers import GPT2Tokenizer, GPT2LMHeadModel, GPT2Model
```

```python
# Set random seed for reproducibility
random.seed(0)
# torch.manual_seed(0)

# Load the GPT-2 medium model and tokenizer
tokenizer = GPT2Tokenizer.from_pretrained('gpt2-medium')
model = GPT2HeadModel.from_pretrained('gpt2-medium')

# Set the device (GPU or CPU)
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model.to(device)
```

```python
# helper functions to return x and y
def get_prompts(data): return L(data).itemgot('prompt')
def get_completions(data): return L(data).itemgot('completion')
```

```python
# Load the training data
corpus = [
    {"prompt":"good for the goose -->","completion":"3"},
    {"prompt":"good -->","completion":"3"},
    {"prompt":"for the goose -->","completion":"2"},
    {"prompt":"for -->","completion":"2"},
    {"prompt":"the goose -->","completion":"2"},
    {"prompt":"quiet , introspective and entertaining independent -->","completion":"4"},
    {"prompt":"quiet , introspective and entertaining -->","completion":"3"},
    {"prompt":"quiet -->","completion":"2"}
]

# Preprocess the data
data = []
for example in corpus:
    prompt = example['prompt']
    completion = example['completion']
    data.append((prompt, completion))

# define tokenizer function
def tokenize(text):
    tokens = tokenizer.encode(text, return_tensors='pt').to(device)
    return tokens

# Define the DataBlock
data_block = DataBlock(
    blocks=(TextBlock(tok_tfm=tokenize), TextBlock(tok_tfm=tokenize)),
    get_x=lambda x: x['prompt'],
    get_y=lambda y: y['completion'],
    splitter=RandomSplitter(valid_pct=0.1, seed=0)
)

# Create the dataloaders
dls = data_block.dataloaders(L(corpus), bs=4)
```

```python
# Set the hyperparameters
num_epochs = 5
learning_rate = 1e-5

# Create the language model learner
learner = language_model_learner(dls, model, config=model.config)

# Fine-tune the model on the corpus
learner.fine_tune(num_epochs, learning_rate)
```
